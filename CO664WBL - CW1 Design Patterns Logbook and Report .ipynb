````python
#MVC_Template_01
#2014 May 23  by Steven Lipton http://makeAppPie.com
#Controller initializing MVC -- simplest version possible.
#Additional annotations by R Mather - Oct 2018

# MVC EXTENDED WITH DELETE BUTTON

# Tkinter is a Python binding to the Tk GUI toolkit/library - RM
from tkinter import *
 
# A Model-View-Controller framework for TKinter.
# Model: Data Structure. Controller can send messages to it, and model can respond to message.
# View : User interface elements. Controller can send messages to it. View can call methods from Controller when an event happens.
# Controller: Ties View and Model together. turns UI responses into chages in data.
 


#Controller: Ties View and Model together.
#       --Performs actions based on View events.
#       --Sends messages to Model and View and gets responses
#       --Has Delegates - RM NOTE: essential for communications between view and model via control
 
class MyController():
    def __init__(self,parent):
        self.parent = parent
        # CRITICAL PART OF THE MVC 
        # RM NOTES:
        # MyController passes a reference to itself (self) when it creates MyModel and MyView objects
        # This allows the MyModel and MyView objects to create 'delegates' (see 'vc' = virtual controller?)
        # The vc objects can then call methods on the MyControl object, thereby allowing the View to update the Model via the Controller 
        self.model = MyModel(self)    # initializes the model
        self.view = MyView(self)  #initializes the view
        # Initialize objects in view
        self.view.setEntry_text('Add to Label') # a non cheat way to do MVC with tkinter control variables
        self.view.setLabel_text('Ready')
    
    #event handlers
    def quitButtonPressed(self):
        self.parent.destroy()
    def addButtonPressed(self):
        self.view.setLabel_text(self.view.entry_text.get())
        # CRITICAL PART OF THE MVC 
        # RM NOTES - In one operation:
        #[1] Get the text from the VIEW;
        #[2] Add it to the MODEL by appending it to the list
        self.model.addToList(self.view.entry_text.get())
        
    # New function to handle the remove button being pressed    
    def removeButtonPressed(self): 
        # Check if the list is not empty
        if self.model.getList():  
            # Remove the last item from the models list and update the view and list
            removedItem = self.model.removeFromList()
            self.view.setLabel_text(f"Removed: {removedItem}")
            self.listChangedDelegate()  
    
    def listChangedDelegate(self):
        #model internally changes and needs to signal a change
        print(self.model.getList())
 
#View : User interface elements.
#       --Controller can send messages to it.
#       --View can call methods from Controller vc when an event happens.
#       --NEVER communicates with Model.
#       --Has setters and getters to communicate with controller
 
class MyView(Frame):
    
    def __init__(self,vc):
        self.frame=Frame()
        self.frame.grid(row = 0,column=0)
        self.vc = vc
        self.entry_text = StringVar()
        self.entry_text.set('nil')
        self.label_text = StringVar()
        self.label_text.set('nil')
        self.loadView()

    def loadView(self):
        quitButton = Button(self.frame,text = 'Quit', command= self.vc.quitButtonPressed).grid(row = 0,column = 0)
        addButton = Button(self.frame,text = "Add", command = self.vc.addButtonPressed).grid(row = 0, column = 1)
        # Added a new remove button to the view and bound it to a new remove function
        removeButton = Button(self.frame, text="Remove", command=self.vc.removeButtonPressed).grid(row=0, column=2)
        entry = Entry(self.frame,textvariable = self.entry_text).grid(row = 1, column = 0, columnspan = 3, sticky = EW)
        label = Label(self.frame,textvariable = self.label_text).grid(row = 2, column = 0, columnspan = 3, sticky = EW)

    def getEntry_text(self):
    #returns a string of the entry text
        return self.entry_text.get()
    def setEntry_text(self,text):
    #sets the entry text given a string
        self.entry_text.set(text)
    def getLabel_text(self):
    #returns a string of the Label text
        return self.label_text.get()
    def setLabel_text(self,text):
    #sets the label text given a string
        self.label_text.set(text)
 
#Model: Data Structure.
#   --Controller can send messages to it, and model can respond to message.
#   --Uses delegates from vc to send messages to the Control of internal change
#   --NEVER communicates with View
#   --Has setters and getters to communicate with Controller
 
class MyModel():
    def __init__(self,vc):
        self.vc = vc
        self.myList = ['duck','duck','goose']
        self.count = 0
#Delegates-- Model would call this on internal change
    def listChanged(self):
        self.vc.listChangedDelegate()
#setters and getters
    def getList(self):
        return self.myList
    def initListWithList(self, aList):
        self.myList = aList
    def addToList(self,item):
        print("returned")
        myList = self.myList
        myList.append(item)
        self.myList=myList
        self.listChanged()
    # New function be called by the button press, and handle removing an item 
    def removeFromList(self):
        # Check if the list is not empty
        if self.myList:  
            # Remove the last item from the list
            removedItem = self.myList.pop()
            # Notify the controller of the change
            self.listChanged()
            # Return the removed item
            return removedItem
        # Return None if the list was empty
        return None  
 
def main():
    # Create a root window from the Tk GUI class
    root = Tk()
    frame = Frame(root,bg='#0555ff' )
    root.title('Hello Penguins')
    app = MyController(root)
    root.mainloop()
 
if __name__ == '__main__':
    main()  
````

````python
import matplotlib.pyplot as plt
import numpy as np


# Nicely abstracted structure by which any model can notify any observer (view) of changes in the model
class Subject(object): #Represents what is being 'observed'

    def __init__(self):
        self._observers = [] # This where references to all the observers are being kept
                             # Note that this is a one-to-many relationship: there will be one subject to be observed by multiple _observers

    def attach(self, observer):
        if observer not in self._observers: #If the observer is not already in the observers list
            self._observers.append(observer) # append the observer to the list

    def detach(self, observer): #Simply remove the observer
        try:
            self._observers.remove(observer)
        except ValueError:
            pass

    def notify(self, modifier=None):
        for observer in self._observers: # For all the observers in the list
            if modifier != observer: # Don't notify the observer who is actually doing the updating 
                observer.update(self) # Alert the observers!
                
# Represents the 'data' for which changes will produce notifications to any registered view/observer objects 
class Model(Subject): # Extends the Subject class

    def __init__(self, name):
        Subject.__init__(self)
        self._name = name # Set the name of the model
        self._labels = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
        self._data = [10,8,6,4,2,1]

    @property #Getter that gets the labels
    def labels(self):
        return self._labels

    @labels.setter #Setter that sets the labels
    def labels(self, labels):
        self._labels = labels
        self.notify() # Notify the observers whenever somebody changes the labels
        
    @property #Getter that gets the data
    def data(self):
        return self._data

    @data.setter #Setter that sets the labels
    def data(self, data):
        self._data = data
        self.notify() # Notify the observers whenever somebody changes the data
        

# This is the 'standard' view/observer which also acts as an 'abstract' class whereby deriving Bar/Chart/Table views override the update() method
# This 'abstracted' layer is always shown in examples but is important to demonstrate potential polymorphic behaviour of update()
class View():
    
    def __init__(self, name=""):
         self._name = name #Set the name of the Viewer

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        print("Generalised Viewer '{}' has: \nName = {}; \nLabels = {}; \nData = {}".format(self._name, subject._name, subject._labels, subject._data))

# Table 'chart' viewer/ConcreteObserver - overrides the update() method
class TableView(View):
    
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        fig = plt.figure(dpi=80)
        ax = fig.add_subplot(1,1,1)
        table_data = list(map(list,zip(subject._labels, subject._data)))
        table = ax.table(cellText=table_data, loc='center')
        table.set_fontsize(14)
        table.scale(1,4)
        ax.axis('off')
        plt.show()

# Bar chart viewer/ConcreteObserver - overrides the update() method
class BarView(View):
        
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        objects = subject._labels
        y_pos = np.arange(len(objects))
        performance = subject._data
        plt.bar(y_pos, performance, align='center', alpha=0.5)
        plt.xticks(y_pos, objects)
        plt.ylabel('Usage')
        plt.title('Programming language usage')
        plt.show()

class PieView(View):
    
    def update(self, subject):  # Alert method that is invoked when the notify() method in a concrete subject is invoked
        # Get the labels and sizes from the subject, using the encapsulated approach of using the property
        labels = subject.labels  
        sizes = subject.data
        explode = (0.1, 0, 0, 0, 0, 0)  # only "explode" the 1st slice
        
        # This function would need additional checks to handle data sets that arent exact;y
        # 6 items in legnth

        fig1, ax1 = plt.subplots()
        ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
        ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()
        


#Let's create our subjects
m1 = Model("Model 1")
m2 = Model("Model 2") # This is never used!

#Let's create our observers
v1 = View("1: standard text viewer")
v2 = TableView("2: table viewer")
v3 = BarView("3: bar chart viewer")
# Create an instance of PieView
v4 = PieView("4: pie chart viewer")

####

#Let's attach our observers to the first model
m1.attach(v1)
m1.attach(v2)
m1.attach(v3)
# Attach the PieView instance to the first model
m1.attach(v4)
####

# Let's just call the notify() method to see all the charts in their unchanged state
m1.notify()

# Now Let's change the properties of our first model
# Change 1 triggers all 4 views and updates their labels
m1.labels = ['C#','PHP','JavaScript','ASP','Python','Smalltalk']
# Change 2 triggers all 4 views and updates their data
m1.data = [1,18,8,60,3,1]
````

````python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np

# Classes that need adapting
class PlateCarree: 
       def __init__(self):
              self.name = "PlateCarree"
       def project_PlateCarree(self):
              ax = plt.axes(projection=ccrs.PlateCarree())
              return ax
    
class InterruptedGoodeHomolosine:
       def __init__(self):
              self.name = "InterruptedGoodeHomolosine"    
       def project_InterruptedGoodeHomolosine(self):
              ax = plt.axes(projection=ccrs.InterruptedGoodeHomolosine())
              return ax

class AlbersEqualArea:
       def __init__(self):
              self.name = "AlbersEqualArea"    
       def project_AlbersEqualArea(self):
              ax = plt.axes(projection=ccrs.AlbersEqualArea())
              return ax   

class Mollweide:
       def __init__(self):
              self.name = "Mollweide"    
       def project_Mollweide(self):
              ax = plt.axes(projection=ccrs.Mollweide())
              return ax
            
# Adapter class
class Adapter:
    # kwargs are used to allow for different objects, with differently named methods that do the same thing
    # to be renamed by the adapter
    def __init__(self, obj, **adapted_method):
        self._object = obj
        self.name = obj.name
        # Add a new dictionary item that maps the adapted_method, to the specific method
        self.__dict__.update(adapted_method)
    
    def __get__attr(self, attr):
        return getattr(self._object, attr)

# Create adapted instances of each class and bind its specific projection method to the Adapter class project method
projections = [
    Adapter(PlateCarree(), project=PlateCarree().project_PlateCarree),
    Adapter(InterruptedGoodeHomolosine(), project=InterruptedGoodeHomolosine().project_InterruptedGoodeHomolosine),
    Adapter(AlbersEqualArea(), project=AlbersEqualArea().project_AlbersEqualArea),
    Adapter(Mollweide(), project=Mollweide().project_Mollweide)
]

# Iterate over each adapted projection class, and perform a set of operations on them
for proj in projections:
    # Use the adapted project method
    ax = proj.project()
    # Perform Cartopys operations on the projection
    ax.stock_img()
    ax.coastlines()
    # Dsiply specific information about each projection
    print(proj.name)
    # Plot the projection
    plt.show()
````

# Mediator Design Pattern #

For Exercise 16 I have chosen to look at the Mediator Design Pattern(Herein referred to as DP). This DP is intended to promote loose coupling between objects and relegate communication responsibilities to a separate communication (mediator) object allowing objects to be ignorant of each others implementation and existence. 

## 1 - Intent ##

The Mediator DP is intended to reduce or remove complex dependencies between multiple objects, it does this by centralising all communication through a Mediator object and encapsulating how the objects interact. Instead of objects referring to each other directly, they interact through this Mediator object. This pattern promotes loose coupling by stopping objects from referring to each other explicitly.

## 2 - Motivation ##

When a system contains many objects, the interaction between those objects can become complex, this makes using, maintaining, and extending the classes difficult. By encapsulating the interaction of the objects into a single Mediator object, the objects only know that the Mediator exists and therefor the communication and control is greatly simplified. The aim of the Mediator is to also reduce the amount of interconnections between objects, lots of interconnections mean that objects have to work with the support of other objects, reducing the ease of changing the systems behaviour. 

## 3 - Structure ##



The structure of the DP would contain:
- **A Mediator:** 
    - Defines an interface for communication with Colleague objects
- **A Concrete Mediator:** 
    - Implements cooperative behaviour by coordingating Colleague objects
    - Knows and maintains its Colleagues
- **A Set of Colleague Classes:**
    - Each Colleague knows its Mediator object
    - Each Colleague communicates with its Mediator, some send and recieve requests, some may just send requests


## 4 - Implementation ##

The implementation of the Mediator DP involves several key considerations:

- 1 -	**Identify interaction points:** Complex interactions need to be identified within the system, a common example of this is user interface elements.
- 2 -	**Defining Mediators role:** The responsibilities of the Mediator need to be clearly defined, it take on a number of roles such as relaying messages, coordinating actions and managing states. The scope of its responsibilities will massively affect the design and complexity of the class. 
- 3 -	**Ensuring loose coupling:** The aim of the Mediator is to have loosely coupled objects, meaning that each component should only interact with the Mediator interface.
- 4 -	**Communication:** The interface for the Mediator needs to have clearly defined communications methods that are consistent and reliable. Depending on its roles, there may be multiple interfaces for achieving different tasks. 
- 5 -	**Complexity of Mediator:** If the Mediator seems to have too many responsibilities or the class is growing to be complex, then splitting the responsibilities up into multiple Mediator classes could be considered.

In the code cell below marked "# Exercise 16 Section 5 - Coded Example #", I have included a basic coded example of a message board. The idea behind this system is that users can post messages and a notification of that message is sent to all users of the messageboard. In this case, the Mediator DP is used to simplify the interactions between the users and the message board, ensuring all communication goes through a centralised object. 

## 6 Evaluation of the Mediator Desing Pattern ##

The Mediator DP is a structural solution that facilitates communication between objects, with the goal of making it easier to maintain, modify and extend systems. The below points summarise my findings from looking into this DP

**Programming language Idioms**
- The Mediator pattern is widely applicable across different programming languages that support OOP. Key features to implement this DP effectively are class inheritance and polymorphism.
- Language features such as interfaces and abstract classes are used to define a common mediator interface, ensuring that the concrete implementations follow consistent interface design. 

**Consequences of using the Design Pattern**
- Reduces coupling between components of a system, each component has far fewer dependencies within the system.
- Centralised control of communication logic simplifies how the components interact, and makes changes to this interaction far easier, as the changes are localised to the Mediator
- Risk of a “God Object”, meaning that if the Mediator is given too many responsibilities, it can actually become hard to maintain and go against the ideology of the DP. 
- Performance considerations need to be accounted for, if the system has a high volume of communication, the Mediator can become a bottleneck and reduce system responsiveness. In the case of this and the “God Object” scenario it could be wise to make multiple Mediators with different responsibilities.

**Appropriate uses**
- **GUI Applications:** In applications were multiple GUI components need to be updated based on user interactions, it can prove beneficial to have this handled by a Mediator, reducing the interconnection between each component. 
- **Distributed System:** In distributed systems the mediator can handle interactions between different parts of the system. Allowing it to act as an event bus between the components.

**Application, Architecture and Implementation Requirements**
- When designing a system around the Mediator DP, it is vital to clearly define the scope of the Mediators responsibilities to avoid it becoming too complex and becoming a “God Object” as mentioned.
- The system should focus on modularity, allowing for the easy addition or removal of components, and changes to their functionality without impacting the other components of the system. 
- Implementations should focus on maintaining a balance between simplifying component interactions, but also not overloading the Mediator with too many responsibilities.

````python
# Exercise 16 Section 5 - Coded Example #

class MessageBoardMediator:
    """
        Mediator class to manage the users and posting of messages
    """
    def __init__(self):
        # Dict to hold users
        self.users = {}
        # List to hold posts
        self.posts = []
        
    def create_user(self, user):
        """
            Adds a new user to the system
        """
        # Add the user to dict of users, and apply the mediator to each new user
        self.users[user.name] = user
        user.mediator = self
    
    def post_message(self, message, user_name):
        """
            Posts a message to the board, and notifies users
        """
        self.posts.append((user_name, message))
        # Iterate through all users and tell them to recieve the message, apart from the sender
        for name, user in self.users.items():
            if name != user_name:
                user.receive_post(message, user_name)
class User:
    """
        Represents a user in the message board system
    """
    def __init__(self, name):
        self.name = name
        # Mediator will be set when the object is added to the MessageBoardMediator object
        self.mediator = None
    
    def post(self, message):
        """
            Post a new message to the message board, via the Mediator
        """
        self.mediator.post_message(message, self.name)
        
    def receive_post(self, message, user_name):
        """
            Recieves a post from the Mediator
        """
        print(f'{self.name} received a post from {user_name}: {message}')
        
        
mediator = MessageBoardMediator()

josh = User("Josh")
steve = User("Steve")
jasmin = User("Jasmin")

mediator.create_user(josh)
mediator.create_user(steve)
mediator.create_user(jasmin)

josh.post("Welcome to the message board")
jasmin.post("Test Message!")
        
        
````

# Prototype Design Pattern #


For exercise 17, I have chosen to look at the Prototype Design Pattern (Herein referred to as DP). This DP lets you copy existing objects without making your code dependent on their classes. It allows for the creation of objects based on the template of an existing object through cloning. 

## 1 - Intent ##

The Prototype DP allows you to specify the kinds of objects to create using a prototypical instance, and create new objects by copying this prototype. It promotes the cloning of objects to avoid the overhead of new instance creation, and allow for dynamic configurations of systems. 

## 2 - Motivation ##

When complex factory class hierarchy are trying to be avoided, the Prototype DP is a good alternative, it allows for the creation of new instances by simply cloning the appropriate prototype. This DP is also useful for dynamic system configurations, in systems that require high degrees of customisation or configurability, the components may not be known until runtime. These can be based on user inputs, configuration files or other parameters. The Prototype DP supports this by allowing systems to specify new classes of objects at runtime and create instances without being bound to specific concrete classes. 

## 3 - Structure ##



The structure of the DP would contain:
- **A Prototype:** 
    - Declares an interface for cloning itself
- **A ConcretePrototype:** 
    - Implements an operation for cloning itself
- **A Client:**
    - Creates a new object by asking a prototype to clone itself


## 4 - Implementation ##

The implementation of the prototype DP involves several key considerations to be effective:

- 1 - **Prototype Registry:** A register of available prototypes needs to be kept, clients wont manage the prototypes themselves, but will store and retrieve them from the registry, the client will ask the registry for a prototype before cloning it.

- 2 - **Cloning:** The decision of deep or shallow cloning needs to be considered for the applications purposes. Shallow cloning will produce objects that share nested objects between the clone and original, whereas a deep copy, provides a completely independent copy.

- 3 - **Initialising Clones:** Some clones may need to, or clients may want to, initialise the clone with different internal values for some or all of its parameters. 

- 4 - **Prototype Modification:** In some cases you may want to alter a prototypes state before cloning it to suit particular needs of the application. 

- 5 - **Prototype Evolution:** As the applications evolve so might the requirements for the prototypes, a mechanism for updating or replacing the prototypes in the registry without affecting clients could be considered. 

- 6 - **Garbage collection:** In languages where resource management and garbage collection are concerns, the objects and copies should be tracked and any unnecessary copies should be disposed of correctly.

## 6 Evaluation of the Prototype Desing Pattern ##

The Prototype DP offers a unique approach to object creation. Focusing on cloning existing instances rather than creating instances of new objects from scratch. The below points summarise my findings from looking into the Prototype DP.

**Programming language Idioms**

- To properly enforce the cloning mechanism, it is important the language supports Abstract classes such as Python with its ABC module and class. This ensures that any concrete prototypes adhere to the structure of the prototypes. 

**Consequences of using the Design Pattern**

````python
# Exercise 17 Section 5 - Coded Example #

from abc import ABC, abstractmethod
import copy

# Making use of the abstract base class 
class DocumentPrototype(ABC):
    """
        Declares an interface for cloning itself.
    """
    @abstractmethod
    def clone(self):
        pass 

class ConcreteDocument(DocumentPrototype):
    """
        Implements an operation for cloning itself
    """
    def __init__(self, title, content):
        self.title = title
        self.content = content
    
    def clone(self):
        """
            Creates a deep coopy of the document, cloning itself
        """
        return copy.deepcopy(self)
    
    def __str__(self):
        return f"Document(title='{self.title}', content='{self.content}')"
    
class DocumentManager:
    """
        A client that manages documents, and can produce them from thje prototypes
    """
    def __init__(self):
        self.document_prototypes = {}
    
    def register_doc_prototype(self, prototype_name, prototype):
        """
            Registers a prototype under a name
        """
        self.document_prototypes[prototype_name] = prototype
    
    def create_document(self, prototype_name, **kwargs):
        """
            Creates a new document by cloning a prototype by name, also takes variable amount of arguments, so 
            that details about the prototype can be adjusted dynamically
        """
        prototype = self.document_prototypes.get(prototype_name)
        # basic checks for making sure the prototype exists
        if not prototype:
            raise ValueError(f"Prototype '{prototype_name}' not found")
        # clone the prorotype using its built in method
        document = prototype.clone()
        # update the self.args with the new arguments provided 
        document.__dict__.update(kwargs)
        return document

# creating instance of the document manager, and registering a basic default document prototype 
document_manager = DocumentManager()
default_doc_prototype = ConcreteDocument("Default Title", "Default Content")
document_manager.register_doc_prototype("default", default_doc_prototype)

# using the registered prototype to create three new document with different attributes 
doc1 = document_manager.create_document("default", title="Project Proposal", content="Project proposal content")
doc2 = document_manager.create_document("default", title="Meeting Notes", content="Notes from the last meeting")
doc3 = document_manager.create_document("default", title="Annual Report", content="Annual report content for 2023")

# printing the documents to show their state
print(doc1)
print(doc2)
print(doc3)  
````
